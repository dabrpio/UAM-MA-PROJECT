```python
from comet import download_model, load_from_checkpoint

def evaluate_comet(src_filename: str, sys_filename: str, refs_filename: str):
    with open(src_filename, 'r') as src_file, \
         open(sys_filename, 'r') as sys_file, \
         open(refs_filename, 'r') as refs_file:
        src = src_file.readlines()
        sys = sys_file.readlines()
        refs = refs_file.readlines()
        data = [
            {"src": src, "mt": mt, "ref": ref}
            for src, mt, ref in zip(src, sys, refs)
        ]

        predictions = model.predict(data, batch_size=8, gpus=1)

        # Output scores
        return sum(predictions["scores"]) / len(predictions["scores"])
```

```python
print(evaluate_comet(src_filename='/content/val.en.txt', refs_filename='/content/val.pl.txt', sys_filename='/content/en-pol/00.txt'))
print(evaluate_comet(src_filename='/content/val.en.txt', refs_filename='/content/val.pl.txt', sys_filename='/content/en-pol/02.txt'))
print(evaluate_comet(src_filename='/content/val.en.txt', refs_filename='/content/val.pl.txt', sys_filename='/content/en-pol/05.txt'))
print(evaluate_comet(src_filename='/content/val.en.txt', refs_filename='/content/val.pl.txt', sys_filename='/content/en-pol/10.txt'))
```
